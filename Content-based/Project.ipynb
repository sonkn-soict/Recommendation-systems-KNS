```python
# thu vien cu phap tru tuong hoa du lieu
import ast
```

```python
#ham chuyen doi doi tuong chi ke thua thuoc tinh 'name'
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_df_cleaned = movies_df_merge.copy()
```

```python
movies_df_cleaned['genres'] =  movies_df_merge['genres'].apply(convert)
movies_df_cleaned['keywords'] = movies_df_merge['keywords'].apply(convert)
```

```python
movies_df_cleaned.head()
```

```python
#ham chon 3 nv chinh
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L    
            
```

```python
movies_df_cleaned['cast'] = movies_df_merge['cast'].apply(convert3)
```

```python
movies_df_cleaned.head()
```

```python
#tim ra giam doc
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies_df_cleaned['crew'] = movies_df_merge['crew'].apply(fetch_director)
```

```python
#thu vien xu li ngon ngu tu nhien
import nltk
```

```python
#thu gon ve dang goc cua tu
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
movies['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
# bien doi thanh danh sach va sap xep theo hang voi key la gi tri cosine
sorted(list(enumerate(similarity[0])),  reverse = True, key = lambda x:x[1])[1:6]
```

```python

def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:6]

    for i in movie_list:
        print(movies.iloc[i[0]].title)
```
